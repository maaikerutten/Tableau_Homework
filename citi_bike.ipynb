### SUMMER (july-aug-sep 2018)



```python
#dataframe for the summer months
summer_df = df_Jul2018.append([df_Aug2018, df_Sep2018])
```

```python
summer_df.shape
```

```python
summer_df = summer_df[['tripduration', 'starttime', 'start station id', 'bikeid', 'gender']]
```

```python
summer_df.head()
```

```python
# Create an empty list 
list_dt = []

# Loop through 'starttime' column, split by 'space' then append into our empty list
for i in summer_df.starttime:
    list_dt.append(i.split(' '))
    
# create 2 columns and unzip the list 
summer_df['startdate'],summer_df['time'] = zip(*list_dt)
```

```python
summer_df.tail()
```

```python
summer_df.dtypes
```

```python
# taking the hours
summer_df.time = summer_df.time.str[0:2]
```

```python
summer_df.head()
```

```python
# groupby object
groupby = summer_df.groupby(summer_df.time)
```

```python
groupby_df = pd.DataFrame()

groupby_df['hourOfDay'] = groupby['time'].first()
groupby_df['count'] = groupby['time'].count().unique()
```

```python
groupby_df.to_csv('summer_hours.csv')
```

```python
df_Dec2018.head()
```

```python
df_Dec2018 = df_Dec2018[['tripduration', 'starttime', 'start station id', 'start station name',
                       'end station id', 'end station name',
                           'bikeid', 'gender']]

# Create an empty list 
list_Dec2018 = []

# Loop through 'starttime' column, split by 'space' then append into our empty list
for i in df_Dec2018.starttime:
    list_Dec2018.append(i.split(' '))
    
# create 2 columns and unzip the list 
df_Dec2018['startdate'],df_Dec2018['time'] = zip(*list_Dec2018)
```

```python
df_Dec2018.head(10)
```

```python
# groupby object
groupby_object_id = df_Dec2018.groupby(df_Dec2018['start station id'])

groupby_startstation_id = pd.DataFrame()

groupby_startstation_id['station_id'] = groupby_object_id['start station id'].first()
groupby_startstation_id['count'] = groupby_object_id['start station id'].count()
groupby_startstation_id['station_name'] = groupby_object_id['start station name'].first()
```

```python
top10_start_df = groupby_startstation_id.sort_values(by='count', ascending=False).head(10)
top10_start_df
```

```python
bottom10_start_df = groupby_startstation_id.sort_values(by='count', ascending=True).head(10)
bottom10_start_df
```

```python
#save top 10 start stations
top10_start_df.to_csv('top10_start_Dec2018.csv')
```

```python
#save bottom 10 start stations
bottom10_start_df.to_csv('bottom10_start_Dec2018.csv')
```

```python
# groupby object
groupby_object_id1 = df_Dec2018.groupby(df_Dec2018['end station id'])

groupby_endstation_id = pd.DataFrame()

groupby_endstation_id['station_id'] = groupby_object_id1['end station id'].first()
groupby_endstation_id['count'] = groupby_object_id1['end station id'].count()
groupby_endstation_id['station_name'] = groupby_object_id1['end station name'].first()
```

```python
top10_end_df = groupby_endstation_id.sort_values(by='count', ascending=False).head(10)
top10_end_df
```

```python
bottom10_end_df = groupby_endstation_id.sort_values(by='count', ascending=True).head(10)
bottom10_end_df
```

```python
#save top 10 end stations 
top10_end_df.to_csv('top10_end_Dec2018.csv')
```

```python
# save bottom 10 end stations
bottom10_df.to_csv('top10_end_Dec2018.csv')
```
