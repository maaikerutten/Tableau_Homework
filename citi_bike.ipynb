### Top 10 starting stations for the month of march

```python
# groupby object
groupby_object_id = df_032019.groupby(df_032019['start station id'])

# groupby manipulation
groupby_sstation_id = pd.DataFrame()

groupby_sstation_id['station_id'] = groupby_object_id['start station id'].first()
groupby_sstation_id['count'] = groupby_object_id['start station id'].count()
groupby_sstation_id['station_name'] = groupby_object_id['start station name'].first()
```

```python
# sorting the values in a descending manner and filtering for the first 10 
top10_start_df = groupby_sstation_id.sort_values(by='count', ascending=False).head(10)
top10_start_df
```

### Bottom 10 for the month of March 2019

```python
# groupby object
groupby_object_id1 = df_032019.groupby(df_032019['end station id'])

# groupby manipulation
groupby_estation_id = pd.DataFrame()

groupby_estation_id['station_id'] = groupby_object_id1['end station id'].first()
groupby_estation_id['count'] = groupby_object_id1['end station id'].count()
groupby_estation_id['station_name'] = groupby_object_id1['end station name'].first()

# sorting the values in a descending manner and filtering for the first 10 
bottom10_df = groupby_estation_id.sort_values(by='count', ascending=True).head(10)
bottom10_df
```
